```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
from keras import backend as k
```

```python
img=image.load_img('E:/image_project/training/0_sarees/1.jpg')
```

```python
plt.imshow(img)
```

```python
cv2.imread('E:/image_project/training/0_sarees/1.jpg').shape
```

```python
cv2.imread('E:/image_project/training/0_sarees/1.jpg')
```

```python
train=ImageDataGenerator(rescale=1/247)
validation=ImageDataGenerator(rescale=1/247)
```

```python
train_dataset=train.flow_from_directory('E:/image_project/training',
                                       target_size=(200,200),
                                       batch_size=10,
                                       class_mode='categorical')
validation_dataset=train.flow_from_directory('E:/image_project/validation',
                                       target_size=(200,200),
                                       batch_size=10,
                                       class_mode='categorical')
```

```python
train_dataset[1]
```

```python
train_dataset.class_indices
```

```python
train_dataset.classes
```

```python
model=tf.keras.models.Sequential([tf.keras.layers.Conv2D(30,(3,3),activation='relu',input_shape=(200,200,3)),
                                 tf.keras.layers.MaxPool2D(2,2),
                                  
                                 tf.keras.layers.Conv2D(60,(3,3),activation='relu'),
                                 tf.keras.layers.MaxPool2D(2,2),
                                  
                                 tf.keras.layers.Conv2D(90,(3,3),activation='relu'),
                                 tf.keras.layers.MaxPool2D(2,2),
                                 
                                 tf.keras.layers.Flatten(),
                                 
                                 tf.keras.layers.Dense(3,activation='softmax')])

```

```python
model.summary()
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
```

```python
model_fit=model.fit_generator(train_dataset,steps_per_epoch=10,epochs=20,validation_data=validation_dataset)
```

```python
model.save_weights('img_class.h5')
```

```python
# serialize model to JSON
model_json = model.to_json()
with open("model.json", "w") as json_file:
    json_file.write(model_json)
```

# Testing the model

```python

import numpy as np
from keras.preprocessing import image
test_image = cv2.imread('E:/image_project/testing/240.jpg')
test_image = cv2.resize(test_image,(200,200))
test_image=test_image/248
test_image = np.expand_dims(test_image, axis = 0)
result = model.predict(test_image)
train_dataset.class_indices
if result[0][0] > 0.5:
    prediction = 'saree'
if result[0][1] > 0.5:
    prediction = 'trouser'
if result[0][2] > 0.5:
    prediction = 'jeans'

print(result)    
print(prediction)



image = cv2.imread('E:/image_project/testing/240.jpg')
image=cv2.resize(image,(200,200))

plt.imshow(image)
plt.xlabel(prediction)
```
